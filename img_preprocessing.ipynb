Amplitude Section of the code

This part pulls the height of the image out

```python
import cv2 as cv
import os

data = os.listdir('amplitude/data')

maxY = []

ColMax = -1
currMax = -1
for file in data:
    im = cv.imread('amplitude/data/' + file)
    for y in range(im.shape[0]):
        if (im[y, 200, 0] > currMax):
            ColMax = y
            currMax = im[y, 200, 0]
    maxY.append(ColMax)

maxY
```

```python
for y in range(len(maxY)):
    maxY[y] = 270-maxY[y]

maxY
```

```python
import matplotlib.pyplot as plt
from matplotlib import spines as sp
import numpy as np

#x = np.array([0,0.0417*1,0.0417*2,0.0417*3,0.0417*4,0.0417*5,0.0417*6,0.0417*7,0.0417*8,0.0417*9])
y = np.array(maxY)

x=np.linspace(0, 0.0417*44, num=45)

#x=np.linspace()
plt.axes().spines['bottom'].set_position(('zero'))
#plt.axes().spines[['left']].set_position('zero')
plt.plot(x,y, 'o', label='data')
plt.plot(x,y)

#.spines[['top', 'right']].set_visible(False)
#plt.axhline(0, color='black')
#plt.axvline(0, color='black')
plt.show()
```

```python
from scipy.interpolate import make_interp_spline
from scipy.interpolate import CubicSpline

#x = np.array([0,0.0417*1,0.0417*2,0.0417*3,0.0417*4,0.0417*5,0.0417*6,0.0417*7,0.0417*8,0.0417*9])
y = np.array(maxY)

x=np.linspace(0, 0.417*45, num=45)

X_Y_Spline = make_interp_spline(x, y)

X_ = np.linspace(x.min(), x.max(), 100)
Y_ = X_Y_Spline(X_)
plt.axes().spines['bottom'].set_position(('zero'))
plt.plot(x,y, 'o', label='data')
plt.plot(X_, Y_)
plt.show()
```
